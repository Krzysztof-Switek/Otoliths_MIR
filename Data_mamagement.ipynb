DIVIDING PHOTOS INTO CATEGORIES ALONG WITH CHANGING THEIR NAMES

```python
import os
import shutil  # Dodane do obsługi kopiowania plików

# Path to the directory with photos
photos_directory_1 = r'C:\Users\Krzychu\OneDrive\Dokumenty\MIR_otolits\RAW_DATA\JesienneWiosenne\JPG'
photos_directory_2 = r'C:\Users\Krzychu\OneDrive\Dokumenty\MIR_otolits\RAW_DATA\WiosenneWiosenneMorskie\JPG'

df = df_1 #change that line for second data frame

# New directory to store copies with modified names
output_directory = r'C:\Users\Krzychu\OneDrive\Dokumenty\MIR_otolits\photos'
print(f"Creating directory: {output_directory}")

# Create the output directory if it doesn't exist
if not os.path.exists(output_directory):
    os.makedirs(output_directory)

# Iterate through files in the photo directory
for filename in os.listdir(photos_directory_1):

    # Check if the file is an image
    if filename.endswith('.JPG') or filename.endswith('.png'):

        # Extract the identifier (e.g., 'A(1)') from the filename to determine the class
        identifier = filename.split('.')[0]

        # Check classification based on the spreadsheet
        row = df[df['Photo_name'] == identifier]
        if not row.empty:
            classification = row['Populacja'].values[0]
            oth_id = row['OTH_ID'].values[0]

            # Create a new filename with modified name
            new_filename = f"oth_{oth_id}_{classification}_{filename}"

            # Create the destination path in the output directory
            destination_path = os.path.join(output_directory, new_filename)

            # Copy the file to the destination path with the modified name
            source_path = os.path.join(photos_directory_1, filename)
            shutil.copyfile(source_path, destination_path)

print("Process completed.")
```
